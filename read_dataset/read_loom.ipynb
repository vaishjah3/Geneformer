```python
import loompy
import matplotlib.pyplot as plt

# Open the .loom file
file_path = 'pancreas_scib.loom'
with loompy.connect(file_path) as ds:
    # Access the data matrix
    data = ds[:, :]

# Visualize the data (e.g., as a heatmap)
plt.figure(figsize=(10, 6))
plt.imshow(data, aspect='auto', cmap='viridis')
plt.xlabel('Genes')
plt.ylabel('Cells')
plt.colorbar(label='Expression level')
plt.title('Expression heatmap')
plt.show()
```
